```python
import os
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.corpus import wordnet as wn
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
folder = "texts"
files = os.listdir(folder)

print(files)
```

```python
def read_file(path):
    with open(path, 'r', encoding='utf-8') as f:
        return f.read()

all_tokens = []

for file in files:
    path = os.path.join(folder, file)
    text = read_file(path)
    
    tokens = word_tokenize(text)
    tokens = [t.lower() for t in tokens]
    
    all_tokens.extend(tokens)

print(len(all_tokens))
```

```python
unique_tokens = set(all_tokens)
print(len(unique_tokens))
```

```python
tagged = pos_tag(unique_tokens)
print(tagged[:20])
```

```python
verbs = [word for word, tag in tagged if tag.startswith('VB')]
verb_set = set(verbs)

print(len(verb_set))
```

## What I noticed

The most common verbs in the Adventure Time scripts were very action-based (like “go,” “say,” “look”). This makes sense for AdventureTime because the dialogu is fast-paced and focused on movement and interaction between characters.

```python
verbs = []

for token in doc:
    if token.pos_ == "VERB":
        verbs.append(token.lemma_.lower())

print(verbs[:20])
```

```python
verb_counts = Counter(verbs)

print(verb_counts.most_common(10))
```

```python
o = open("verbFreq.txt", "w", encoding="utf-8")

for verb, count in verb_counts.items():
    o.write(verb + "\t" + str(count) + "\n")

o.close()
```
