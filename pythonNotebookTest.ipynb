3b. Prepare a code block that depends on running the previous code block and then reads from each directory of files and calculates what's needed to indicate **lexical diversity**.
    Running the cell that you prepare should reveal the lexical diversity of each set of files. *(3 pts)*

```python
import os
from nltk.tokenize import word_tokenize

season1_path = "texts/season1"
season10_path = "texts/season10"

def load_corpus(folder_path):
    text = ""
    for file in os.listdir(folder_path):
        if file.endswith(".txt"):
            with open(os.path.join(folder_path, file), "r", encoding="utf-8") as f:
                text += f.read().lower() + " "
    return text

season1_text = load_corpus(season1_path)
season10_text = load_corpus(season10_path)

def lexical_diversity(text):
    tokens = word_tokenize(text)
    if len(tokens) == 0:
        return 0
    return len(set(tokens)) / len(tokens)

season1_ld = lexical_diversity(season1_text)
season10_ld = lexical_diversity(season10_text)

print("Season 1 Lexical Diversity:", season1_ld)
print("Season 10 Lexical Diversity:", season10_ld)
```

```python
import spacy

def extract_verbs(text):
    doc = nlp(text)
    return [token.text for token in doc if token.pos_ == "VERB"]

season1_verbs = extract_verbs(season1_text)
season10_verbs = extract_verbs(season10_text)

print("Season 1 Verbs (sample):", season1_verbs[:20])
print("Season 10 Verbs (sample):", season10_verbs[:20])
```

6. Develop the code in the previous block by copying it into the next code block, and here:
   * Develop code to find out about the top 15 most ambiguous words in your adjective or verb token list, and
   * Write your output to a neatly formatted text file for reading. The text output file should provide a clear heading to provide a title for each corpus you're comparing. *(5 pts)*

```python
from nltk.corpus import wordnet as wn

def ambiguity_score(word):
    return len(wn.synsets(word))

def top_ambiguous_words(word_list, n=15):
    candidates = list(set(word_list))
    scored = [(word, ambiguity_score(word)) for word in candidates]
    scored = [item for item in scored if item[1] > 1]
    scored.sort(key=lambda x: x[1], reverse=True)
    return scored[:n]

season1_words = [w.lower() for w in season1_verbs]
season10_words = [w.lower() for w in season10_verbs]

season1_ambiguous = top_ambiguous_words(season1_words, 15)
season10_ambiguous = top_ambiguous_words(season10_words, 15)

output_text = ""

output_text += "SEASON 1 - TOP 15 MOST AMBIGUOUS VERBS\n"
output_text += "-" * 50 + "\n"
for word, score in season1_ambiguous:
    output_text += f"{word}: {score} senses\n"

output_text += "\n\nSEASON 10 - TOP 15 MOST AMBIGUOUS VERBS\n"
output_text += "-" * 50 + "\n"
for word, score in season10_ambiguous:
    output_text += f"{word}: {score} senses\n"

with open("ambiguous_words_output.txt", "w", encoding="utf-8") as f:
    f.write(output_text)

print("Output file created: ambiguous_words_output.txt")
```
