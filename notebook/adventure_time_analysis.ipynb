```python


import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords


nltk.download('punkt')
nltk.download('stopwords')


with open("ATE1.txt", "r", encoding="utf-8") as f:
    text1 = f.read()

with open("ATE2.txt", "r", encoding="utf-8") as f:
    text2 = f.read()


tokens1 = word_tokenize(text1)
tokens2 = word_tokenize(text2)


def lexical_diversity(tokens):
    return len(set(tokens)) / len(tokens)


print("Script 1 Diversity:", lexical_diversity(tokens1))
print("Script 2 Diversity:", lexical_diversity(tokens2))


fdist1 = FreqDist(tokens1)
fdist1.plot(20)


stop_words = set(stopwords.words('english'))

filtered1 = [w for w in tokens1 if w.lower() not in stop_words]
filtered2 = [w for w in tokens2 if w.lower() not in stop_words]


FreqDist(filtered1).plot(20)
FreqDist(filtered2).plot(20)
```
